```python
# Execute diretamente no Jupyter
import pandas as pd
import numpy as np

# Carregar dados
df = pd.read_csv('kubernetes_ml_dataset.csv')

print("="*80)
print("üîç VALIDA√á√ÉO R√ÅPIDA DO DATASET")
print("="*80)

# 1. Informa√ß√µes b√°sicas
print(f"\nüìä Dimens√µes: {df.shape[0]} linhas x {df.shape[1]} colunas")

# 2. Primeiras linhas
print("\nüëÄ Primeiras 5 linhas:")
display(df.head())

# 3. Informa√ß√µes das colunas
print("\nüìã Informa√ß√µes das colunas:")
df.info()

# 4. Valores nulos
print("\n‚ùì Valores nulos:")
null_counts = df.isnull().sum()
null_pct = (null_counts / len(df)) * 100
null_df = pd.DataFrame({
    'Nulos': null_counts,
    'Percentual': null_pct
}).sort_values('Nulos', ascending=False)
display(null_df[null_df['Nulos'] > 0])

if null_df['Nulos'].sum() == 0:
    print("‚úÖ Nenhum valor nulo encontrado!")

# 5. Duplicatas
duplicates = df.duplicated().sum()
print(f"\nüîÑ Duplicatas: {duplicates} ({(duplicates/len(df)*100):.2f}%)")

# 6. Estat√≠sticas
print("\nüìà Estat√≠sticas descritivas:")
display(df.describe())

# 7. Tipos de dados
print("\nüî§ Tipos de dados:")
display(df.dtypes)

# 8. Colunas √∫nicas
print("\nüè∑Ô∏è  Valores √∫nicos por coluna:")
for col in df.columns:
    unique_count = df[col].nunique()
    print(f"   {col}: {unique_count} valores √∫nicos")

# 9. Score de qualidade
null_score = 100 - (null_df['Nulos'].sum() / (len(df) * len(df.columns)) * 100 * 2)
dup_score = 100 - (duplicates / len(df) * 100 * 3)
quality_score = (null_score + dup_score) / 2

print(f"\n{'='*80}")
print(f"üéØ SCORE DE QUALIDADE: {quality_score:.1f}/100")
print(f"{'='*80}")

if quality_score >= 90:
    print("‚úÖ EXCELENTE - Dataset de alta qualidade")
elif quality_score >= 75:
    print("üëç BOM - Dataset com qualidade aceit√°vel")
else:
    print("‚ö†Ô∏è  REGULAR - Dataset precisa de melhorias")
```
